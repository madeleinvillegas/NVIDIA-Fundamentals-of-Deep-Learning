# Selecting Data in a DataFrame

Our goal is to understand further on how to select a subset of data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

We’re going to use a new dataset here, to demonstrate how to deal with larger datasets. This is a subset of the of 311 service requests from [NYC Open Data](https://nycopendata.socrata.com/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9).

```python
complaints = pd.read_csv('https://media.githubusercontent.com/media/johnanthonyjose/python-data-analysis/master/Module3_Pandas/311-service-requests.csv')
```

Depending on your pandas version, you might see an error like "DtypeWarning: Columns (8) have mixed types". This means that it’s encountered a problem reading in our data. In this case it almost certainly means that it has columns where some of the entries are strings and some are integers.

For now we’re going to ignore it and hope we don’t run into a problem, but in the long run we’d need to investigate this warning.

## Selecting Columns and Rows


We can start first by visualizing the existing columns names

```python
complaints.columns
```

```python
complaints['Complaint Type']
```

If you care to observe the row-wise, then

```python
complaints[:5]
```

`[:5]` indicates you only want to see the first five rows

What if you want data from 7:13?

```python
complaints[7:13]
```

What if you want row 7-13 of 'Complain Type'?

We can use the `loc` method.

```python
complaints.loc[7:13,'Complaint Type']
```

Alternatively,

```python
complaints['Complaint Type'][7:13]
```

# Selecting Multiple Columns and Multiple Rows

It is possible to select some rows and some muli-columns

```python
complaints.loc[7:13,['Complaint Type','Descriptor']]
```

Alternatively, if you want samples from 7:13, 151:200 on both columns

```python
idx1 = (complaints.index < 13) & (complaints.index >= 7)
idx2 = (complaints.index < 200) & (complaints.index >= 100)

idx = idx1 | idx2

complaints.loc[idx,['Complaint Type','Descriptor']]
```
